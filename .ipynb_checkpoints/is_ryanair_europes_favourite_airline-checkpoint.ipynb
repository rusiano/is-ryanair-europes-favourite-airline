### 2.1.  Load Airlines

```python
airlines_df = pd.read_csv('data/airlines.csv', header=None, na_values=['\\N', '-', 'NAN'])
airlines_df.columns = ['id', 'name', 'alias', 'iata', 'icao', 'callsign', 'country', 'active']

# drop:
# - airlines that do not have a IATA code
# - airlines that are no longer active
airlines_df = airlines_df.dropna(subset=['iata'])
airlines_df = airlines_df[airlines_df.active == 'Y'].copy()

# standardize all the names belonging to the same IATA code by taking the most used one
airlines_df.name = airlines_df.iata.replace(
    dict(airlines_df.groupby('iata').name.apply(lambda series: series.value_counts().index[0])))


airlines_df.head()
```

### 2.2. Load Airports

```python
airports_df = pd.read_csv('data/airports.csv', header=None, na_values=['\\N', '-'])
airports_df.columns = ['id', 'name', 'city', 'country', 'iata', 'icao', 'latitude', 'longitude', 'altitude', 'timezone', 'dst', 'tz_timezone', 'type', 'data_source']


airports_df.head()
```

### 2.2.1. Select European Airports

```python
european_codes_df = pd.read_csv('data/european_iatas_df.csv', header=None)
european_codes_df.columns = ['city', 'airport_name', 'iata']


european_codes_df.head()
```

Use the dataset to select only European airports:

```python
european_airports_df = airports_df[airports_df.iata.isin(european_codes_df.iata.str.strip().tolist())].copy()


european_airports_df.head()
```

### 2.3. Routes

```python
routes_df = pd.read_csv('data/routes.csv', header=None, na_values=['\\N', '-'])
routes_df.columns = ['airline', 'airline_id', 'src', 'src_id', 'dest', 'dest_id', 'codeshare', 'stops', 'equipment']
```

According to docs *src* and *dest* may contain either IATA or ICAO code, depending whether the code has 3 or 4 letters, respectively. Let's check if they are indeed mixed or all of one type.

```python
_src_iata = routes_df.src.apply(lambda code: code if len(code) == 3 else np.nan)
_src_icao = routes_df.src.apply(lambda code: code if len(code) == 4 else np.nan)

_dest_iata = routes_df.dest.apply(lambda code: code if len(code) == 3 else np.nan)
_dest_icao = routes_df.dest.apply(lambda code: code if len(code) == 4 else np.nan)

print('Is "src" field containing only IATA codes? {}'.format(_src_iata.isna().sum() == 0))
print('Is "dest" field containing only IATA codes? {}'.format(_dest_iata.isna().sum() == 0))
```

This mean that *src* and *dest* can be actually considered IATA codes. For better understanding let's rename the two fields.

```python
routes_df.rename(columns={'src': 'src_iata', 'dest': 'dest_iata'})


routes_df.head()
```

### 2.3.1 Select European Routes

Let's select only routes between two European airports:

```python
european_iatas = european_airports_df.iata.unique()
european_routes_df = routes_df[routes_df.src_iata.isin(european_iatas) 
                               & routes_df.dest_iata.isin(european_iatas)].copy()

# drop non-relevant fields
european_routes_df.drop(columns=['airline', 'codeshare', 'stops', 'equipment'], inplace=True)

# join table with previous dataset
european_routes_df = (
    european_routes_df.copy()
    .join(
        airlines_df.set_index('id').name.rename('airline_name'), on='airline_id')
    .join(
        european_codes_df.set_index('iata').city.rename('src_city'), on='src_iata')
    .join(
        european_codes_df.set_index('iata').city.rename('dest_city'), on='dest_iata')
    .join(
        airports_df.set_index('iata').country.rename('src_country'), on='src_iata')
    .join(
        airports_df.set_index('iata').country.rename('dest_country'), on='dest_iata'))

# drop routes for which the airline is missing
european_routes_df.dropna(subset=['airline_name'], inplace=True)

# add a unique code for the undirected connection (i.e., same for flight from A to B and from B to A)
european_routes_df['connection_iata'] = (
    european_routes_df[['src_iata', 'dest_iata']].copy()
    .apply(lambda r: '_'.join(sorted(r)), axis=1))

european_routes_df.head()
```

## Exclusive Connections

We can now compute how many connections are served exclusively by one and only one airline, and count the number of exclusive connections per airline.

```python
# the exclusive connections are those served by only one airline
exclusive_connections_df = connections_df[connections_df.n_airlines == 1].copy()

# extract the connected iatas from the index
exclusive_connections_df[['iata_A', 'iata_B']] = (
    exclusive_connections_df
    .index.to_series()
    .apply(lambda s: pd.Series(s.split('_'))))

# use the iatas to append the correposnding cities
exclusive_connections_df = (
    exclusive_connections_df
    .join(
        airports_df.set_index('iata').city.rename('city_A'),
        on='iata_A')
    .join(
        airports_df.set_index('iata').city.rename('city_B'),
        on='iata_B'))
    
exclusive_connections_df.head()
```

```python
fig, (ax,ax1) = plt.subplots(1,2,facecolor='w', figsize=(12,6))

n_exclusive_connections = (
    exclusive_connections_df
    .groupby('airlines_names').size()
    .sort_values(ascending=False))

(n_exclusive_connections.iloc[:25][::-1].plot(
    kind='barh', 
    color=['k' if name != 'Ryanair' else 'dodgerblue' for name in n_exclusive_connections.iloc[:25][::-1].index.tolist()], 
    ax=ax))


exclusive_connections_ratio = ((n_exclusive_connections / n_connections)
 .dropna()
 .reindex(n_connections.sort_values(ascending=False).iloc[:25].index.tolist(), fill_value=0)
).sort_values()

exclusive_connections_ratio.plot(
    kind='barh', 
    color=['k' if name != 'Ryanair' else 'dodgerblue' for name in exclusive_connections_ratio.index.tolist()], 
    ax=ax1)

ax.set_title('Top 25 Airlines\nfor Number of Exclusive Connections')
ax.set_ylabel('Airline')
ax.set_xlabel('# Exclusive Connections')

ax1.set_title('Exclusive Connections Ratio\n(for Top 25 Airlines by Number of Connections in Europe)')
ax1.set_ylabel('')
ax1.set_xlabel('Exclusive Connections Ratio')

plt.tight_layout()
plt.show()
```

## Presence

It is also possible to compute some statistics about the presence of an airline in the various airports (and countries).
For example, in how many airports is an airline present? Or assuming that the presence of an airline in an airport can be computed as the fraction of flights from/to the given airport owned by the given company over the total amount of flights of the airport, what's the average or median presence of an airline across the airports it serves?

```python
airport_presence_stats = (
    european_routes_df_lite.copy()
    .groupby(['src_iata', 'airline_name']).size()
    .groupby('src_iata').apply(lambda g: g / g.sum())
    .reset_index(0,drop=True)
    .groupby('airline_name')
    .agg({'avg_presence': 'mean', 'mdn_presence': 'median', 'n_served_airports': 'count'})
    .applymap(lambda v: round(v,2)))

airport_presence_stats.head()
```

```python
fig, ax = plt.subplots(figsize=(6,6), facecolor='w')

top25_n_serv_airports = (
    airport_presence_stats
    .n_served_airports
    .sort_values()
    .iloc[-25:])

(top25_n_serv_airports
 .plot(kind='barh', 
       color=['k' if name != 'Ryanair' else 'dodgerblue' for name in top25_n_serv_airports.index.tolist()], 
       ax=ax))

ax.set_title('Top 25 Airlines\nby Number of Airports Served in Europe')
ax.set_xlabel('# Served Airports')
ax.set_ylabel('Airline')

plt.show()
```

```python
fig, (ax,ax1) = plt.subplots(1,2, figsize=(12,6), facecolor='w')
    
top25_mdn_presence = (
    airport_presence_stats
    .mdn_presence
    .sort_values()
    .loc[n_connections.sort_values(ascending=False).iloc[:25].index]
    .sort_values())

top25_avg_presence = (
    airport_presence_stats
    .avg_presence
    .sort_values()
    .loc[n_connections.sort_values(ascending=False).iloc[:25].index]
    .sort_values())
    
(top25_mdn_presence
 .plot(kind='barh', 
       color=['k' if name != 'Ryanair' else 'dodgerblue' for name in top25_mdn_presence.index.tolist()], 
       ax=ax))

(top25_avg_presence
 .plot(kind='barh', 
       color=['k' if name != 'Ryanair' else 'dodgerblue' for name in top25_avg_presence.index.tolist()],
       ax=ax1))

ax.set_title('Median Presence per Served Airport in Europe\nfor Top 25 Airlines by Number of Connections')
ax.set_xlabel('MDN airport presence')
ax.set_ylabel('Airline')

ax1.set_title('Average Presence per Served Airport in Europe\nfor Top 25 Airlines by Number of Connections')
ax1.set_ylabel('')
ax1.set_xlabel('AVG airport presence')

plt.tight_layout()
plt.show()
```

Ryanair shows to have on average roughly half of the flights of the airports it serves. This reduces the options for travellers to chose another airline...

```python
country_presence_stats = (
    european_routes_df_lite.copy()
    .groupby(['src_country', 'airline_name']).size()
    .groupby('src_country').apply(lambda g: g / g.sum())
    .reset_index(0,drop=True)
    .groupby('airline_name')
    .agg({'avg_presence': 'mean', 'mdn_presence': 'median', 'n_served_airports': 'count'})
    .applymap(lambda v: round(v,2)))

country_presence_stats.head()
```

```python
fig, (ax,ax1) = plt.subplots(1,2, figsize=(12,6), facecolor='w')
    
top25_mdn_presence = (
    country_presence_stats
    .mdn_presence
    .sort_values()
    .loc[n_connections.sort_values(ascending=False).iloc[:25].index]
    .sort_values())

top25_avg_presence = (
    country_presence_stats
    .avg_presence
    .sort_values()
    .loc[n_connections.sort_values(ascending=False).iloc[:25].index]
    .sort_values())
    
(top25_mdn_presence
 .plot(kind='barh', 
       color=['k' if name != 'Ryanair' else 'dodgerblue' for name in top25_mdn_presence.index.tolist()], 
       ax=ax))

(top25_avg_presence
 .plot(kind='barh', 
       color=['k' if name != 'Ryanair' else 'dodgerblue' for name in top25_avg_presence.index.tolist()],
       ax=ax1))

ax.set_title('Median Presence per Served Country in Europe\nfor Top 25 Airlines by Number of Connections')   
ax.set_xlabel('MDN country presence')
ax.set_ylabel('Airline')

ax1.set_title('Average Presence per Served Country in Europe\nfor Top 25 Airlines by Number of Connections')
ax1.set_ylabel('')
ax1.set_xlabel('AVG country presence')

plt.tight_layout()
plt.show()
```
